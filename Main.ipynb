```python
import requests

url = "https://www.cars24.com/article/10-most-expensive-cars-in-the-world/"

response = requests.get(url)
print(response.text)
```

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(response.text, 'html.parser')
name = soup.find_all('span',class_='ml-3 font-semibold')

unique_text = set()
car =[]

for span in name:
    text = span.text.strip()
    #print(text)
    if text in unique_text:
        continue
    unique_text.add(text)
    parts = text.split('|')
    #print(parts)

    if(len(parts)>=2):
        car_name = parts[0].split('.',1)[1].strip()
        indian_price = parts[1].strip()
        dollar_price = parts[2].strip()

        car.append(
            {
             'name' : car_name,
             'INR' : indian_price,
             'Dollar':dollar_price
        }
        )
    df = pd.DataFrame(car)
    df.to_excel('TopCar.xlsx', index=False)




for i in car:
    print(f"Car:{i['name']},INR:{i['INR']},Dollar:{i['Dollar']}")

```
